###-ALGORITMO-I- Caja

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta
from faker import Faker
fake_es = Faker('es_AR') # O 'es_MX', 'es_CO', etc. según tu preferencia de español regional
# Inicialización

np.random.seed(42)
random.seed(42)
Faker.seed(42)


descripciones_gasto = [
    "Pago de servicios",
    "Compra de suministros",
    "Mantenimiento de equipos",
    "Gastos de representación",
    "Alquiler de oficina",
    "Pago a proveedores",
    "Reparación de maquinaria",
    "Transporte y logística",
    "Capacitación del personal",
    "Honorarios profesionales"
]
# Parámetros
num_registros = 30
responsables = [fake_es.name() for _ in range(10)]
tipos_transaccion = ['Venta', 'Gasto']
metodos_pago = ['Efectivo', 'Tarjeta de Débito', 'Tarjeta de Crédito', 'Transferencia']
categorias_productos = ['Electrónica', 'Alimentos', 'Ropa', 'Accesorios', 'Juguetes']
saldo = 50000

# Lista de registros
registros = []

for i in range(num_registros):
    # Fecha aleatoria dentro de los últimos 6 meses
    fecha_hora_transaccion = fake_es.date_time_between(start_date='-6M', end_date='now')
    tipo_transaccion = random.choice(tipos_transaccion)
    monto = round(random.uniform(1000, 15000), 2)

    # Ajuste del saldo
    if tipo_transaccion == 'Venta':
        saldo += monto
    else:
        saldo -= monto

    registro = {
        'id_transaccion': i + 1,
        'fecha_hora': fecha_hora_transaccion.strftime('%Y-%m-%d %H:%M:%S'),
        'tipo_transaccion': tipo_transaccion,
        'metodo_pago': random.choice(metodos_pago),
        'monto': monto,
        'saldo_acumulado': round(saldo, 2),
        'cajero_id': fake_es.random_int(min=1, max=10),
        'numero_ticket': fake_es.unique.bothify(text='TK-########'),
        'cliente_id': fake_es.random_int(min=1000, max=9999) if random.random() > 0.3 else None,
        'producto_categoria': random.choice(categorias_productos) if tipo_transaccion == 'Venta' else None,
    'descripcion': random.choice(descripciones_gasto) if tipo_transaccion == 'Gasto' else None,
        'responsable': random.choice(responsables)
    }

    registros.append(registro)

# Convertir a DataFrame
df_transacciones = pd.DataFrame(registros)

# Ordenar por fecha
df_transacciones.sort_values(by='fecha_hora', inplace=True)
df_transacciones.reset_index(drop=True, inplace=True)

# Mostrar muestra
##print(df_transacciones.head())

# Guardar si deseas
df_transacciones.to_csv('auditoria_caja.csv', index=False)
df_transacciones
```

### Algoritmo II-Conciliacion Bancaria

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# --- Generar datos simulados ---
fake = Faker('es_AR')
np.random.seed(123)
random.seed(123)

num_transacciones = 30
tipos_transaccion = ['Débito', 'Crédito']
conceptos_banco = ['Transferencia recibida', 'Transferencia enviada', 'Pago proveedor', 'Cobro cliente', 'Intereses', 'Comisión bancaria', 'Depósito efectivo']
conceptos_contables = ['Pago proveedor', 'Cobro cliente', 'Intereses ganados', 'Comisión pagada', 'Depósito registrado']

def generar_fecha():
    return fake.date_between(start_date='-30d', end_date='today')

movimientos_banco = []
movimientos_contables = []

for i in range(num_transacciones):
    fecha = generar_fecha()
    tipo = random.choice(tipos_transaccion)
    monto = round(random.uniform(500, 10000), 2)
    concepto_banco = random.choice(conceptos_banco)
    concepto_contable = random.choice(conceptos_contables)

    movimientos_banco.append({
        'id_banco': i + 1,
        'fecha': fecha,
        'tipo': tipo,
        'concepto_banco': concepto_banco,
        'monto_banco': monto,
        'referencia_banco': fake.unique.bothify('BNK-########')
    })

    # 90% de chance de tener contrapartida contable
    if random.random() > 0.1:
        monto_contable = monto if random.random() > 0.2 else round(monto + random.uniform(-100, 100), 2)
        movimientos_contables.append({
            'id_contable': i + 1,
            'fecha': fecha,
            'tipo': tipo,
            'concepto_contable': concepto_contable,
            'monto_contable': monto_contable,
            'referencia_contable': fake.unique.bothify('CNT-########')
        })

df_banco = pd.DataFrame(movimientos_banco)
df_contabilidad = pd.DataFrame(movimientos_contables)

# --- Conciliación aproximada ---
# Merge tolerando diferencias de hasta $50 y por fecha/tipo
tolerancia = 50

# Expandir el dataset contable para permitir combinación múltiple (simulación)
df_merged = pd.merge(df_banco, df_contabilidad,
                     on=['fecha', 'tipo'],
                     how='outer', suffixes=('_banco', '_contable'))

# Calcular diferencia
df_merged['diferencia_monto'] = np.abs(df_merged['monto_banco'] - df_merged['monto_contable'])

# Marcar conciliación
df_merged['conciliado'] = df_merged['diferencia_monto'] <= tolerancia

# Llenar NaNs en campos faltantes por outer merge
df_merged['conciliado'] = df_merged['conciliado'].fillna(False)

# Mostrar resultados
df_merged[['fecha', 'tipo', 'monto_banco', 'monto_contable', 'diferencia_monto', 'conciliado']].head(10)

# Guardar si lo deseas
df_merged.to_csv('conciliacion_bancaria.csv', index=False)
df_merged
```

### Algoritmo III- Cuentas a cobrar

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Inicializar Faker y semillas
np.random.seed(123)
random.seed(123)
fake = Faker('es_AR')
Faker.seed(123)

# Parámetros
num_clientes = 30
num_facturas = 60
tipos_cliente = ['Empresa', 'Individuo']
estados_posibles = ['Pagada', 'Pendiente', 'Vencida', 'Parcialmente Pagada']
condiciones_pago = [0, 15, 30, 60]  # en días

# Generar clientes simulados
clientes = []
for i in range(num_clientes):
    clientes.append({
        'cliente_id': 1000 + i,
        'nombre': fake.company() if random.random() > 0.3 else fake.name(),
        'tipo': random.choice(tipos_cliente),
        'cuit': fake.unique.bothify(text='##-########-#')
    })

# Generar facturas (cuentas a cobrar)
facturas = []
for i in range(num_facturas):
    cliente = random.choice(clientes)
    fecha_emision = fake.date_between(start_date='-90d', end_date='today')
    dias_pago = random.choice(condiciones_pago)
    fecha_vencimiento = fecha_emision + timedelta(days=dias_pago)
    estado = random.choices(estados_posibles, weights=[0.4, 0.3, 0.2, 0.1])[0]
    monto_total = round(random.uniform(5000, 150000), 2)

    # Determinar monto pagado y fecha de pago según el estado
    if estado == 'Pagada':
        monto_pagado = monto_total
        fecha_pago = fecha_emision if dias_pago == 0 else fecha_emision + timedelta(days=random.randint(0, dias_pago))
    elif estado == 'Parcialmente Pagada':
        monto_pagado = round(monto_total * random.uniform(0.1, 0.9), 2)
        fecha_pago = fecha_emision if dias_pago == 0 else fecha_emision + timedelta(days=random.randint(0, dias_pago))
    else:
        monto_pagado = 0.0
        fecha_pago = None

    saldo_pendiente = round(monto_total - monto_pagado, 2)

    facturas.append({
        'factura_id': f'F-{10000 + i}',
        'cliente_id': cliente['cliente_id'],
        'fecha_emision': fecha_emision,
        'fecha_vencimiento': fecha_vencimiento,
        'fecha_pago': fecha_pago,
        'estado': estado,
        'monto_total': monto_total,
        'monto_pagado': monto_pagado,
        'saldo_pendiente': saldo_pendiente,
        'condicion_pago_dias': dias_pago
    })

# Convertir a DataFrame
df_facturas = pd.DataFrame(facturas)
df_clientes = pd.DataFrame(clientes)

# Unir para obtener listado final
df_cuentas_cobrar = pd.merge(df_facturas, df_clientes, on='cliente_id')

# Ordenar por fecha de vencimiento
df_cuentas_cobrar.sort_values(by='fecha_vencimiento', inplace=True)

# Mostrar muestra
df_cuentas_cobrar

# Guardar a CSV si lo necesitás
df_cuentas_cobrar.to_csv("listado_cuentas_a_cobrar.csv", index=False)
df_cuentas_cobrar
```

### algoritmo IV- Gastos pagados por adelantado

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Inicialización
np.random.seed(42)
random.seed(42)
fake = Faker('es_AR')
Faker.seed(42)

# Parámetros
num_registros = 30
tipos_gasto = ['Alquiler', 'Seguro', 'Publicidad', 'Licencias de software', 'Mantenimiento']
proveedores = [fake.company() for _ in range(10)]

# Generar datos
prepagos = []

for i in range(num_registros):
    proveedor = random.choice(proveedores)
    tipo = random.choice(tipos_gasto)
    fecha_pago = fake.date_between(start_date='-60d', end_date='today')

    # Duración del servicio prepago (1 a 12 meses)
    duracion_meses = random.choice([1, 3, 6, 12])
    fecha_inicio = fecha_pago + timedelta(days=random.randint(0, 15))
    fecha_fin = fecha_inicio + pd.DateOffset(months=duracion_meses)

    # Monto proporcional al tiempo
    monto_total = round(random.uniform(5000, 200000), 2)
    monto_mensual = round(monto_total / duracion_meses, 2)

    prepagos.append({
        'id_prepago': f'PP-{1000 + i}',
        'tipo_gasto': tipo,
        'proveedor': proveedor,
        'fecha_pago': fecha_pago,
        'fecha_inicio_servicio': fecha_inicio,
        'fecha_fin_servicio': fecha_fin.date(),
        'duracion_meses': duracion_meses,
        'monto_total': monto_total,
        'monto_mensual': monto_mensual,
        'moneda': 'ARS',
        'estado': 'Activo' if fecha_fin > datetime.today() else 'Consumido'
    })

# Convertir a DataFrame
df_prepagos = pd.DataFrame(prepagos)

# Guardar si lo necesitás
df_prepagos.to_csv("gastos_pagados_adelantado.csv", index=False)
df_prepagos
```

### Algoritmo V- Inventario de materia prima

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# Inicialización de semillas para reproducibilidad
np.random.seed(42)
random.seed(42)

# Definición de parámetros
num_items = 30 # Número de registros a generar
categorias = ['Metales', 'Plásticos', 'Químicos', 'Textiles', 'Madera', 'Papel', 'Vidrio']
unidades = ['kg', 'litros', 'm3', 'unidades', 'rollos', 'metros']

# Lista de descripciones comunes para materias primas
descripciones_comunes = [
    "Cable de Cobre", "Resina Epoxi", "Cloruro de Calcio", "Tela de Algodón", "Madera de Cedro",
    "Papel Reciclado", "Vidrio Templado", "Chapa de Acero", "Polipropileno", "Sulfato de Amonio",
    "Hilo de Poliéster", "Roble Blanco", "Cartón Prensado", "Botella PET", "Aluminio Extruido",
    "PVC Flexible", "Ácido Nítrico", "Lana Sintética", "Aglomerado MDF", "Pulpa de Celulosa",
    "Fibra Óptica", "Bronce Fundido", "Nylon Grado Industrial", "Disolvente Orgánico", "Seda Natural",
    "Abeto Seco", "Film de Embalaje", "Cristal de Cuarzo", "Plomo Puro", "Espuma de Poliestireno"
]

# Asegurarse de que el número de ítems no exceda las descripciones disponibles
if num_items > len(descripciones_comunes):
    # Si num_items es mayor, se pueden repetir descripciones
    descripciones_final = random.choices(descripciones_comunes, k=num_items)
else:
    descripciones_final = random.sample(descripciones_comunes, k=num_items)


# Lista para almacenar los datos del inventario
inventario_data = []

# Generación de datos para cada registro
for i in range(num_items):
    id_material = f'MP-{1000 + i}'
    descripcion = descripciones_final[i] # Usar una descripción de la lista
    categoria = random.choice(categorias)
    unidad = random.choice(unidades)

    cantidad_stock = round(random.uniform(10, 1000), 2)
    punto_reposicion = round(random.uniform(5, 150), 2)
    costo_unitario = round(random.uniform(1, 500), 2)
    valor_total = round(cantidad_stock * costo_unitario, 2)

    # Generar fechas dentro de un rango realista
    # La fecha de entrada será hasta hace 1 año
    fecha_base = datetime.now() - timedelta(days=random.randint(30, 365))
    fecha_ultima_entrada = fecha_base.strftime('%Y-%m-%d')

    # Asegurar que la fecha de salida sea posterior a la de entrada (dentro de los 60 días siguientes)
    fecha_ultima_salida = (fecha_base + timedelta(days=random.randint(1, 60))).strftime('%Y-%m-%d')

    inventario_data.append({
        'id_material': id_material,
        'descripcion': descripcion,
        'categoria': categoria,
        'unidad_medida': unidad,
        'cantidad_stock': cantidad_stock,
        'punto_reposicion': punto_reposicion,
        'costo_unitario': costo_unitario,
        'valor_total': valor_total,
        'fecha_ultima_entrada': fecha_ultima_entrada,
        'fecha_ultima_salida': fecha_ultima_salida
    })

# Crear el DataFrame de Pandas
df_inventario_mp = pd.DataFrame(inventario_data)

# Exportar el DataFrame a un archivo CSV
# El parámetro index=False evita que Pandas escriba el índice del DataFrame como una columna en el CSV.
df_inventario_mp.to_csv("inventario_materia_prima.csv", index=False)
df_inventario_mp
```

### Algoritmo VII - Productos Terminados

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Inicialización
np.random.seed(42)
random.seed(42)
fake = Faker('es_AR')
Faker.seed(42)

# Parámetros para la generación de datos
num_registros_wip = 50 # Número base de procesos para simular (puede generar menos productos terminados)
productos_base = ['Silla Estándar', 'Mesa de Comedor', 'Estantería Modular', 'Armario Doble', 'Puerta Interior']
etapas_produccion = ['Corte', 'Ensamblado', 'Soldadura', 'Pintura', 'Control de Calidad']
lineas_produccion = ['Línea A', 'Línea B', 'Línea C']

# --- Generación de Productos en Proceso (WIP) ---
# Esta sección es necesaria para tener una base de la cual "terminar" productos.
# No se exportará si solo se pide el código de productos terminados, pero es el origen de los mismos.
productos_en_proceso_temp = []

for i in range(num_registros_wip):
    producto = random.choice(productos_base)
    lote = f'L-{fake.random_int(min=1000, max=9999)}'
    etapa = random.choice(etapas_produccion)
    linea = random.choice(lineas_produccion)
    cantidad_total = random.randint(50, 200)
    fecha_inicio = fake.date_between(start_date='-60d', end_date='-7d') # Rango de inicio más amplio
    duracion_estimada = random.randint(3, 20)  # días estimados para terminar
    fecha_estim_termino = fecha_inicio + timedelta(days=duracion_estimada)
    avance_porcentaje = random.randint(70, 100) # Más probabilidad de estar cerca de terminado
    estado = 'En proceso' if avance_porcentaje < 100 else 'Terminado'

    # Calcular cantidades en proceso y terminadas para este ciclo
    cantidad_en_proceso_calculada = int(cantidad_total * ((100 - avance_porcentaje) / 100))
    cantidad_terminada_calculada = cantidad_total - cantidad_en_proceso_calculada

    productos_en_proceso_temp.append({
        'id_proceso': f'WIP-{1000 + i}',
        'producto': producto,
        'lote': lote,
        'etapa_actual': etapa,
        'linea_produccion': linea,
        'cantidad_total_lote': cantidad_total,
        'cantidad_en_proceso_wip': cantidad_en_proceso_calculada,
        'cantidad_terminada_wip': cantidad_terminada_calculada, # Cantidad que ya está "lista" del lote WIP
        'fecha_inicio_proceso': fecha_inicio,
        'fecha_estim_termino_proceso': fecha_estim_termino,
        'avance_porcentaje_wip': avance_porcentaje,
        'estado_wip': estado
    })

df_wip_temp = pd.DataFrame(productos_en_proceso_temp)

# --- Generación de Productos Terminados ---
productos_terminados_final = []

for _, row in df_wip_temp.iterrows():
    # Solo consideramos productos que tienen una cantidad terminada (ya sea parcial o total)
    if row['cantidad_terminada_wip'] > 0:
        # La fecha de término real podría ser ligeramente posterior a la estimada
        fecha_termino_real = row['fecha_estim_termino_proceso'] + timedelta(days=random.randint(0, 5))

        # Estado del producto terminado (generalmente "Almacén" o "Disponible")
        estado_pt = 'Almacén' if row['avance_porcentaje_wip'] == 100 else 'Parcialmente en Almacén'

        productos_terminados_final.append({
            'id_producto_terminado': f'PT-{row["id_proceso"].split("-")[1]}', # ID único para el producto terminado
            'producto': row['producto'],
            'lote_produccion': row['lote'],
            'linea_produccion': row['linea_produccion'],
            'cantidad_terminada': row['cantidad_terminada_wip'],
            'fecha_termino_produccion': fecha_termino_real.strftime('%Y-%m-%d'),
            'estado_almacen': estado_pt, # Estado dentro del almacén o listo para envío
            'costo_produccion_unitario': round(random.uniform(100, 1500), 2), # Costo unitario estimado del PT
            'valor_venta_unitario': round(random.uniform(150, 2500), 2) # Precio de venta unitario
        })

df_productos_terminados = pd.DataFrame(productos_terminados_final)

# Exportar el DataFrame de productos terminados a un archivo CSV
nombre_archivo_csv = 'productos_terminados.csv'
df_productos_terminados.to_csv(nombre_archivo_csv, index=False)
df_productos_terminados
```

### Algoritmo XIV- cuentas por pagar

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# --- Configuración para la generación de datos ---
np.random.seed(42) # Para que los resultados sean reproducibles
num_registros = 30 # Cantidad de facturas a generar

# Fecha de referencia para simular vencimientos y estados
# Usamos la fecha actual (Julio 10, 2025) como punto de referencia para los vencimientos.
fecha_actual_referencia = datetime(2025, 7, 10)

# --- Generación de datos aleatorios para cada columna ---
ids_factura = [f'INV-{i:04d}' for i in range(num_registros)]
proveedores = [f'Proveedor_{i}' for i in random.choices(range(1, 21), k=num_registros)] # 20 proveedores únicos
montos = [round(random.uniform(100, 75000), 2) for _ in range(num_registros)] # Montos en pesos o dólares simulados
monedas = random.choices(['USD', 'ARS', 'EUR'], weights=[0.5, 0.4, 0.1], k=num_registros)
estados_posibles = ['Pendiente', 'Pagada', 'Vencida']

fechas_emision = []
fechas_vencimiento = []
estados_final = []

for _ in range(num_registros):
    # Fechas de emisión: últimos 2 años
    emision = fecha_actual_referencia - timedelta(days=random.randint(10, 365 * 2))
    vencimiento_base = emision + timedelta(days=random.randint(5, 120)) # Vence entre 5 y 120 días después de emisión

    # Simular el estado y ajustar fechas/estados según la lógica
    estado_simulado = random.choices(estados_posibles, weights=[0.65, 0.25, 0.10], k=1)[0] # Más pendientes, algunos pagados, algunos vencidos

    if estado_simulado == 'Pagada':
        # Si está pagada, la fecha de vencimiento puede ser el pasado o una fecha de pago cercana a la emisión
        vencimiento = emision + timedelta(days=random.randint(1, 30))
        # Para que sea 'Pagada' pero su vencimiento original podría haber sido futuro
        if vencimiento_base > fecha_actual_referencia:
            vencimiento = vencimiento_base # Si vencía a futuro, la consideramos pagada antes
        else:
            vencimiento = vencimiento_base # Si ya venció, se pagó tarde

    elif estado_simulado == 'Vencida':
        # Si está vencida, la fecha de vencimiento DEBE ser anterior a la fecha actual de referencia
        vencimiento = fecha_actual_referencia - timedelta(days=random.randint(1, 180)) # Vencida hace 1 a 180 días
        # Asegurarse de que la fecha de emisión sea anterior al vencimiento
        if emision >= vencimiento:
            emision = vencimiento - timedelta(days=random.randint(5, 60)) # Emisión antes del vencimiento

    else: # Pendiente
        # Si está pendiente, la fecha de vencimiento debe ser futura respecto a la fecha actual de referencia
        vencimiento = fecha_actual_referencia + timedelta(days=random.randint(1, 90)) # Vence en los próximos 90 días
        # Asegurarse de que la emisión sea antes del vencimiento
        if emision >= vencimiento:
             emision = vencimiento - timedelta(days=random.randint(5, 60))


    fechas_emision.append(emision)
    fechas_vencimiento.append(vencimiento)
    estados_final.append(estado_simulado)


# --- Creación del DataFrame ---
df_cuentas_pagar = pd.DataFrame({
    'id_factura': ids_factura,
    'proveedor': proveedores,
    'fecha_emision': fechas_emision,
    'fecha_vencimiento': fechas_vencimiento,
    'monto': montos,
    'moneda': monedas,
    'estado': estados_final
})

# Asegurarse de que las columnas de fecha sean tipo datetime después de la creación
df_cuentas_pagar['fecha_emision'] = pd.to_datetime(df_cuentas_pagar['fecha_emision'])
df_cuentas_pagar['fecha_vencimiento'] = pd.to_datetime(df_cuentas_pagar['fecha_vencimiento'])

# --- Guardar el DataFrame en un archivo CSV ---
nombre_archivo = 'cuentas_por_pagar.csv'
df_cuentas_pagar.to_csv(nombre_archivo, index=False)

df_cuentas_pagar
```

**### Algoritmo XVII-Anticipos de clientes**

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# --- Configuración para la generación de datos ---
np.random.seed(42) # Para que los resultados sean reproducibles
num_anticipos = 30 # Cantidad de registros de anticipos a generar

# Fecha de referencia para simular fechas de anticipo y aplicación
fecha_actual_referencia = datetime(2025, 7, 10)

# Listas de valores posibles para las columnas
clientes = [f'Cliente_{i}' for i in range(1, 31)] # 30 clientes únicos
monedas = ['USD', 'ARS', 'EUR']
servicios_productos = [
    'Desarrollo de Software', 'Consultoría IT', 'Servicio de Mantenimiento',
    'Licencia Anual', 'Capacitación', 'Diseño Web', 'Soporte Técnico Premium',
    'Implementación de ERP', 'Auditoría Financiera', 'Asesoría Legal'
]
estados_aplicacion = ['Pendiente de aplicación', 'Parcialmente aplicado', 'Totalmente aplicado']

data = []

for i in range(num_anticipos):
    id_cliente = random.choice(clientes)
    fecha_ant = fecha_actual_referencia - timedelta(days=random.randint(10, 365 * 2)) # Anticipos de los últimos 2 años
    monto = round(random.uniform(500, 25000), 2)
    moneda = random.choice(monedas)
    servicio = random.choice(servicios_productos)

    estado_aplic = random.choices(estados_aplicacion, weights=[0.5, 0.3, 0.2], k=1)[0]

    monto_aplic = 0.0
    fecha_cierre = np.nan # Inicialmente nulo

    if estado_aplic == 'Totalmente aplicado':
        monto_aplic = monto
        fecha_cierre = fecha_ant + timedelta(days=random.randint(1, 90)) # Se aplica en los 3 meses siguientes
        # Asegurarse que la fecha de cierre no sea futura si ya está aplicado
        if fecha_cierre > fecha_actual_referencia:
            fecha_cierre = fecha_actual_referencia - timedelta(days=random.randint(1, 30))
            if fecha_cierre < fecha_ant: # Asegurar que el cierre es después del anticipo
                fecha_cierre = fecha_ant + timedelta(days=random.randint(1, 5))
    elif estado_aplic == 'Parcialmente aplicado':
        monto_aplic = round(random.uniform(0.1, 0.9) * monto, 2)
        fecha_cierre = np.nan # No hay fecha de cierre total
    else: # Pendiente de aplicación
        monto_aplic = 0.0
        fecha_cierre = np.nan

    data.append({
        'id_anticipo': f'ANT-{i:04d}',
        'id_cliente': id_cliente,
        'fecha_anticipo': fecha_ant,
        'monto_anticipo': monto,
        'moneda': moneda,
        'servicio_producto_asociado': servicio,
        'estado_aplicacion': estado_aplic,
        'monto_aplicado': monto_aplic,
        'monto_pendiente_aplicar': round(monto - monto_aplic, 2), # Nueva columna calculada
        'fecha_cierre_aplicacion': fecha_cierre # NaN si no está totalmente aplicado
    })

# --- Creación del DataFrame ---
df_anticipos_clientes = pd.DataFrame(data)

# Asegurarse de que las columnas de fecha sean tipo datetime
df_anticipos_clientes['fecha_anticipo'] = pd.to_datetime(df_anticipos_clientes['fecha_anticipo'])
df_anticipos_clientes['fecha_cierre_aplicacion'] = pd.to_datetime(df_anticipos_clientes['fecha_cierre_aplicacion'])


# --- Guardar el DataFrame en un archivo CSV ---
nombre_archivo = 'anticipos_clientes.csv'
df_anticipos_clientes.to_csv(nombre_archivo, index=False)
df_anticipos_clientes
```

### Algoritmo XVIII- Dividendos a Pagar

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# --- Configuración para la generación de datos ---
np.random.seed(42) # Para reproducibilidad
num_dividendos = 30 # Cantidad de registros de dividendos a generar

# Fecha de referencia (hoy) para simular fechas
fecha_actual_referencia = datetime(2025, 7, 10)

# Listas de valores posibles para las columnas
tipos_accionista = ['Local', 'Extranjero']
accionistas_locales = [f'Accionista_L{i}' for i in range(1, 21)] # 20 locales
accionistas_extranjeros = [f'Accionista_E{i}' for i in range(1, 11)] # 10 extranjeros
monedas = ['ARS', 'USD']
estados_pago = ['Pendiente', 'Pagado', 'Retenido'] # Retenido para casos como impuestos

data = []

for i in range(num_dividendos):
    tipo_acc = random.choices(tipos_accionista, weights=[0.7, 0.3], k=1)[0] # Más accionistas locales

    if tipo_acc == 'Local':
        accionista_id = random.choice(accionistas_locales)
        moneda_div = 'ARS' if random.random() < 0.8 else 'USD' # Locales mayormente en ARS
    else:
        accionista_id = random.choice(accionistas_extranjeros)
        moneda_div = 'USD' # Extranjeros siempre en USD para este ejemplo

    fecha_decl = fecha_actual_referencia - timedelta(days=random.randint(30, 365)) # Declarado en el último año
    fecha_pago_est = fecha_decl + timedelta(days=random.randint(15, 90)) # Pago estimado entre 15 y 90 días después

    monto_por_accion = round(random.uniform(0.5, 5.0), 2) # Monto de dividendo por acción
    cantidad_acciones = random.randint(100, 10000) # Cantidad de acciones por accionista
    monto_total_dividendo = round(monto_por_accion * cantidad_acciones, 2)

    estado = random.choices(estados_pago, weights=[0.6, 0.3, 0.1], k=1)[0]
    fecha_pago_real = np.nan # Inicialmente nulo

    if estado == 'Pagado':
        fecha_pago_real = fecha_pago_est - timedelta(days=random.randint(-5, 5)) # Pagado cerca de la fecha estimada
        # Asegurarse que la fecha de pago no sea futura si el estado es 'Pagado'
        if fecha_pago_real > fecha_actual_referencia:
            fecha_pago_real = fecha_actual_referencia - timedelta(days=random.randint(1, 30))
            if fecha_pago_real < fecha_decl: # Pago real no puede ser antes de la declaración
                 fecha_pago_real = fecha_decl + timedelta(days=random.randint(1, 7))

    elif estado == 'Retenido':
        # Dividendo declarado pero retenido (ej. por impuestos o verificación)
        pass # Fecha de pago real sigue siendo NaN

    else: # Pendiente
        # Asegurarse que la fecha de pago estimada sea futura si está pendiente
        if fecha_pago_est <= fecha_actual_referencia:
            fecha_pago_est = fecha_actual_referencia + timedelta(days=random.randint(5, 60))


    data.append({
        'id_dividendo': f'DIV-{i:04d}',
        'tipo_accionista': tipo_acc,
        'accionista': accionista_id,
        'fecha_declaracion': fecha_decl,
        'fecha_pago_estimada': fecha_pago_est,
        'monto_por_accion': monto_por_accion,
        'cantidad_acciones': cantidad_acciones,
        'monto_total_dividendo': monto_total_dividendo,
        'moneda': moneda_div,
        'estado_pago': estado,
        'fecha_pago_real': fecha_pago_real # Será NaN si no está pagado
    })

# --- Creación del DataFrame ---
df_dividendos = pd.DataFrame(data)

# Asegurarse de que las columnas de fecha sean tipo datetime
df_dividendos['fecha_declaracion'] = pd.to_datetime(df_dividendos['fecha_declaracion'])
df_dividendos['fecha_pago_estimada'] = pd.to_datetime(df_dividendos['fecha_pago_estimada'])
df_dividendos['fecha_pago_real'] = pd.to_datetime(df_dividendos['fecha_pago_real'])


# --- Guardar el DataFrame en un archivo CSV ---
nombre_archivo = 'dividendos_a_pagar.csv'
df_dividendos.to_csv(nombre_archivo, index=False)

df_dividendos
```

### Algoritmo XIX-Otras Deudas

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# --- Configuración para la generación de datos ---
np.random.seed(42) # Para que los resultados sean reproducibles
num_deudas = 30 # Cantidad de registros de deudas a generar

# Fecha de referencia (hoy) para simular fechas
fecha_actual_referencia = datetime(2025, 7, 10)

# Listas de valores posibles para las columnas
tipos_deuda = ['Préstamo Bancario', 'Deuda Comercial', 'Arrendamiento Financiero', 'Préstamo de Accionista']
acreedores = [f'Acreedor_{i}' for i in range(1, 21)] # 20 acreedores únicos
monedas = ['ARS', 'USD']
estados_deuda = ['Activa', 'Cancelada', 'Vencida']

data = []

for i in range(num_deudas):
    tipo = random.choice(tipos_deuda)
    acreedor = random.choice(acreedores)
    moneda = random.choice(monedas)

    fecha_orig = fecha_actual_referencia - timedelta(days=random.randint(30, 365 * 5)) # Origen en los últimos 5 años
    fecha_venc = fecha_orig + timedelta(days=random.randint(90, 365 * random.randint(1, 5))) # Vencimiento entre 3 meses y 5 años desde origen

    monto_orig = round(random.uniform(100000, 10000000), 2)

    tasa_interes = round(random.uniform(0.05, 0.45), 4) # Tasa anual, entre 5% y 45%

    estado = random.choices(estados_deuda, weights=[0.7, 0.2, 0.1], k=1)[0]

    saldo_pend = monto_orig

    if estado == 'Cancelada':
        saldo_pend = 0.0
        # Asegurarse que la fecha de vencimiento sea anterior a la fecha actual si está cancelada y se asumió a tiempo
        if fecha_venc > fecha_actual_referencia:
            fecha_venc = fecha_orig + timedelta(days=random.randint(30, 180)) # Vencimiento en el pasado
            if fecha_venc > fecha_actual_referencia: # Asegurar que sea pasado
                fecha_venc = fecha_actual_referencia - timedelta(days=random.randint(1, 30))

    elif estado == 'Vencida':
        # Si está vencida, el saldo pendiente es aleatorio (no 0) y la fecha de vencimiento debe ser pasada
        saldo_pend = round(random.uniform(0.1, 0.9) * monto_orig, 2)
        fecha_venc = fecha_actual_referencia - timedelta(days=random.randint(1, 180)) # Vencida hace 1 a 180 días
        # Asegurarse que la fecha de origen sea anterior al vencimiento
        if fecha_orig >= fecha_venc:
            fecha_orig = fecha_venc - timedelta(days=random.randint(30, 365))
    else: # Activa
        # Si está activa, el saldo pendiente es un porcentaje del original y la fecha de vencimiento es futura o muy cercana
        saldo_porcentaje = random.uniform(0.1, 1.0) # Puede quedar algo pendiente o todo
        saldo_pend = round(monto_orig * saldo_porcentaje, 2)
        if fecha_venc <= fecha_actual_referencia: # Si venció o vence hoy, ajustar a futuro
            fecha_venc = fecha_actual_referencia + timedelta(days=random.randint(1, 365*3))


    data.append({
        'id_deuda': f'DEU-{i:04d}',
        'tipo_deuda': tipo,
        'acreedor': acreedor,
        'fecha_origen': fecha_orig,
        'fecha_vencimiento': fecha_venc,
        'monto_original': monto_orig,
        'saldo_pendiente': saldo_pend,
        'moneda': moneda,
        'tasa_interes_anual': tasa_interes,
        'estado_deuda': estado
    })

# --- Creación del DataFrame ---
df_otras_deudas = pd.DataFrame(data)

# Asegurarse de que las columnas de fecha sean tipo datetime
df_otras_deudas['fecha_origen'] = pd.to_datetime(df_otras_deudas['fecha_origen'])
df_otras_deudas['fecha_vencimiento'] = pd.to_datetime(df_otras_deudas['fecha_vencimiento'])


# --- Guardar el DataFrame en un archivo CSV ---
nombre_archivo = 'otras_deudas.csv'
df_otras_deudas.to_csv(nombre_archivo, index=False)


df_otras_deudas
```

### Algoritmo  XX-Previsiones

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# --- Configuración para la generación de datos ---
np.random.seed(42) # Para que los resultados sean reproducibles
num_previsiones = 30 # Cantidad de registros de previsiones a generar

# Fecha de referencia (hoy) para simular fechas
fecha_actual_referencia = datetime(2025, 7, 10)

# Listas de valores posibles para las columnas
tipos_prevision = ['Garantías', 'Litigios', 'Cobranzas Dudosas', 'Reestructuración', 'Devoluciones de Ventas', 'Desmantelamiento']
probabilidades = ['Alta', 'Media', 'Baja']
estados_prevision = ['Activa', 'Utilizada', 'Revertida', 'Ajustada']

data = []

for i in range(num_previsiones):
    tipo = random.choice(tipos_prevision)

    desc_breve = f"Previsión por {tipo} - Evento {i+1}"

    # Aseguramos que los argumentos de randint sean enteros
    fecha_creacion = fecha_actual_referencia - timedelta(days=random.randint(int(30), int(365 * 3))) # Creada en los últimos 3 años
    monto_estimado = round(random.uniform(100000.0, 5000000.0), 2) # Aseguramos que sean floats para uniform
    probabilidad = random.choice(probabilidades)

    estado = random.choices(estados_prevision, weights=[0.6, 0.2, 0.1, 0.1], k=1)[0]

    # Aseguramos que los argumentos de randint sean enteros
    fecha_ult_rev = fecha_creacion + timedelta(days=random.randint(int(15), int(365)))
    if fecha_ult_rev > fecha_actual_referencia:
        fecha_ult_rev = fecha_actual_referencia - timedelta(days=random.randint(int(0), int(30))) # Si la revisión es futura, ajustarla al pasado reciente

    fecha_est_utilizacion = np.nan # Inicialmente nulo

    if estado == 'Activa':
        fecha_est_utilizacion = fecha_actual_referencia + timedelta(days=random.randint(int(30), int(365 * 2))) # Utilización futura
    elif estado == 'Utilizada':
        monto_estimado = round(monto_estimado * random.uniform(0.8, 1.2), 2) # Podría ser diferente al estimado
        fecha_est_utilizacion = fecha_creacion + timedelta(days=random.randint(int(30), int(365 * 1.5))) # Utilizada 1 mes a 1.5 años después de creación
        if fecha_est_utilizacion > fecha_actual_referencia:
            fecha_est_utilizacion = fecha_actual_referencia - timedelta(days=random.randint(int(1), int(60))) # Asegurar que esté en el pasado
    elif estado == 'Revertida':
        fecha_est_utilizacion = fecha_creacion + timedelta(days=random.randint(int(30), int(365 * 1.5))) # Revertida en un período similar
        if fecha_est_utilizacion > fecha_actual_referencia:
            fecha_est_utilizacion = fecha_actual_referencia - timedelta(days=random.randint(int(1), int(60)))
        monto_estimado = 0.0 # O un monto pequeño si fue revertida parcialmente
    elif estado == 'Ajustada':
        monto_estimado = round(monto_estimado * random.uniform(0.5, 1.5), 2) # Monto puede haber cambiado significativamente
        fecha_est_utilizacion = fecha_actual_referencia + timedelta(days=random.randint(int(30), int(365))) # Aún activa con nuevo monto

    data.append({
        'id_prevision': f'PREV-{i:04d}',
        'tipo_prevision': tipo,
        'descripcion_breve': desc_breve,
        'fecha_creacion': fecha_creacion,
        'monto_estimado_ars': monto_estimado,
        'probabilidad_ocurrencia': probabilidad,
        'estado_actual': estado,
        'fecha_ultima_revision': fecha_ult_rev,
        'fecha_estimada_utilizacion': fecha_est_utilizacion
    })

# --- Creación del DataFrame ---
df_previsiones = pd.DataFrame(data)

# Asegurarse de que las columnas de fecha sean tipo datetime
df_previsiones['fecha_creacion'] = pd.to_datetime(df_previsiones['fecha_creacion'])
df_previsiones['fecha_ultima_revision'] = pd.to_datetime(df_previsiones['fecha_ultima_revision'])
df_previsiones['fecha_estimada_utilizacion'] = pd.to_datetime(df_previsiones['fecha_estimada_utilizacion'])


# --- Guardar el DataFrame en un archivo CSV ---
nombre_archivo = 'previsiones.csv'
df_previsiones.to_csv(nombre_archivo, index=False)

df_previsiones
```
