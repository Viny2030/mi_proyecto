```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# --- Generar datos simulados ---
fake = Faker('es_AR')
np.random.seed(123)
random.seed(123)

num_transacciones = 30
tipos_transaccion = ['Débito', 'Crédito']
conceptos_banco = ['Transferencia recibida', 'Transferencia enviada', 'Pago proveedor', 'Cobro cliente', 'Intereses', 'Comisión bancaria', 'Depósito efectivo']
conceptos_contables = ['Pago proveedor', 'Cobro cliente', 'Intereses ganados', 'Comisión pagada', 'Depósito registrado']

def generar_fecha():
    return fake.date_between(start_date='-30d', end_date='today')

movimientos_banco = []
movimientos_contables = []

for i in range(num_transacciones):
    fecha = generar_fecha()
    tipo = random.choice(tipos_transaccion)
    monto = round(random.uniform(500, 10000), 2)
    concepto_banco = random.choice(conceptos_banco)
    concepto_contable = random.choice(conceptos_contables)

    movimientos_banco.append({
        'id_banco': i + 1,
        'fecha': fecha,
        'tipo': tipo,
        'concepto_banco': concepto_banco,
        'monto_banco': monto,
        'referencia_banco': fake.unique.bothify('BNK-########')
    })

    # 90% de chance de tener contrapartida contable
    if random.random() > 0.1:
        monto_contable = monto if random.random() > 0.2 else round(monto + random.uniform(-100, 100), 2)
        movimientos_contables.append({
            'id_contable': i + 1,
            'fecha': fecha,
            'tipo': tipo,
            'concepto_contable': concepto_contable,
            'monto_contable': monto_contable,
            'referencia_contable': fake.unique.bothify('CNT-########')
        })

df_banco = pd.DataFrame(movimientos_banco)
df_contabilidad = pd.DataFrame(movimientos_contables)

# --- Conciliación aproximada ---
# Merge tolerando diferencias de hasta $50 y por fecha/tipo
tolerancia = 50

# Expandir el dataset contable para permitir combinación múltiple (simulación)
df_merged = pd.merge(df_banco, df_contabilidad,
                     on=['fecha', 'tipo'],
                     how='outer', suffixes=('_banco', '_contable'))

# Calcular diferencia
df_merged['diferencia_monto'] = np.abs(df_merged['monto_banco'] - df_merged['monto_contable'])

# Marcar conciliación
df_merged['conciliado'] = df_merged['diferencia_monto'] <= tolerancia

# Llenar NaNs en campos faltantes por outer merge
df_merged['conciliado'] = df_merged['conciliado'].fillna(False)

# Mostrar resultados
df_merged[['fecha', 'tipo', 'monto_banco', 'monto_contable', 'diferencia_monto', 'conciliado']].head(10)

# Guardar si lo deseas
df_merged.to_csv('conciliacion_bancaria.csv', index=False)
df_merged
```

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Inicializar Faker y semillas
np.random.seed(123)
random.seed(123)
fake = Faker('es_AR')
Faker.seed(123)

# Parámetros
num_clientes = 30
num_facturas = 100
tipos_cliente = ['Empresa', 'Individuo']
estados_posibles = ['Pagada', 'Pendiente', 'Vencida', 'Parcialmente Pagada']
condiciones_pago = [0, 15, 30, 60]  # en días

# Generar clientes simulados
clientes = []
for i in range(num_clientes):
    clientes.append({
        'cliente_id': 1000 + i,
        'nombre': fake.company() if random.random() > 0.3 else fake.name(),
        'tipo': random.choice(tipos_cliente),
        'cuit': fake.unique.bothify(text='##-########-#')
    })

# Generar facturas (cuentas a cobrar)
facturas = []
for i in range(num_facturas):
    cliente = random.choice(clientes)
    fecha_emision = fake.date_between(start_date='-90d', end_date='today')
    dias_pago = random.choice(condiciones_pago)
    fecha_vencimiento = fecha_emision + timedelta(days=dias_pago)
    estado = random.choices(estados_posibles, weights=[0.4, 0.3, 0.2, 0.1])[0]
    monto_total = round(random.uniform(5000, 150000), 2)

    # Determinar monto pagado y fecha de pago según el estado
    if estado == 'Pagada':
        monto_pagado = monto_total
        fecha_pago = fecha_emision if dias_pago == 0 else fecha_emision + timedelta(days=random.randint(0, dias_pago))
    elif estado == 'Parcialmente Pagada':
        monto_pagado = round(monto_total * random.uniform(0.1, 0.9), 2)
        fecha_pago = fecha_emision if dias_pago == 0 else fecha_emision + timedelta(days=random.randint(0, dias_pago))
    else:
        monto_pagado = 0.0
        fecha_pago = None

    saldo_pendiente = round(monto_total - monto_pagado, 2)

    facturas.append({
        'factura_id': f'F-{10000 + i}',
        'cliente_id': cliente['cliente_id'],
        'fecha_emision': fecha_emision,
        'fecha_vencimiento': fecha_vencimiento,
        'fecha_pago': fecha_pago,
        'estado': estado,
        'monto_total': monto_total,
        'monto_pagado': monto_pagado,
        'saldo_pendiente': saldo_pendiente,
        'condicion_pago_dias': dias_pago
    })

# Convertir a DataFrame
df_facturas = pd.DataFrame(facturas)
df_clientes = pd.DataFrame(clientes)

# Unir para obtener listado final
df_cuentas_cobrar = pd.merge(df_facturas, df_clientes, on='cliente_id')

# Ordenar por fecha de vencimiento
df_cuentas_cobrar.sort_values(by='fecha_vencimiento', inplace=True)

# Mostrar muestra
df_cuentas_cobrar

# Guardar a CSV si lo necesitás
df_cuentas_cobrar.to_csv("listado_cuentas_a_cobrar.csv", index=False)
df_cuentas_cobrar
```

### Algoritmo VII

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Inicialización
np.random.seed(42)
random.seed(42)
fake = Faker('es_AR')
Faker.seed(42)

# Parámetros
num_registros = 30
productos = ['Silla', 'Mesa', 'Estantería', 'Armario', 'Puerta']
etapas_produccion = ['Corte', 'Ensamblado', 'Soldadura', 'Pintura', 'Control de calidad']
lineas_produccion = ['Línea A', 'Línea B', 'Línea C']

# Generar Productos en Proceso (WIP)
productos_proceso = []

for i in range(num_registros):
    producto = random.choice(productos)
    lote = f'L-{fake.random_int(min=1000, max=9999)}'
    etapa = random.choice(etapas_produccion)
    linea = random.choice(lineas_produccion)
    cantidad_total = random.randint(50, 200)
    fecha_inicio = fake.date_between(start_date='-30d', end_date='-1d')
    duracion_estimada = random.randint(5, 15)  # días estimados para terminar
    fecha_estim_termino = fecha_inicio + timedelta(days=duracion_estimada)
    avance_porcentaje = random.randint(10, 100)
    estado = 'En proceso' if avance_porcentaje < 100 else 'Terminado'
    cantidad_en_proceso = int(cantidad_total * (avance_porcentaje / 100))
    cantidad_terminada = cantidad_total - cantidad_en_proceso if estado == 'En proceso' else cantidad_total

    productos_proceso.append({
        'id_proceso': f'WIP-{1000 + i}',
        'producto': producto,
        'lote': lote,
        'etapa_actual': etapa,
        'linea_produccion': linea,
        'cantidad_total': cantidad_total,
        'cantidad_en_proceso': cantidad_en_proceso,
        'fecha_inicio': fecha_inicio,
        'fecha_estim_termino': fecha_estim_termino,
        'avance_porcentaje': avance_porcentaje,
        'estado': estado
    })

df_wip = pd.DataFrame(productos_proceso)

# Generar Productos Terminados basados en WIP terminados o parcial completados
productos_terminados = []

for _, row in df_wip.iterrows():
    # Si está terminado, todo va a terminados
    if row['estado'] == 'Terminado':
        productos_terminados.append({
            'id_terminado': f'PT-{row["id_proceso"].split("-")[1]}',
            'producto': row['producto'],
            'lote': row['lote'],
            'linea_produccion': row['linea_produccion'],
            'cantidad_terminada': row['cantidad_total'],
            'fecha_termino': row['fecha_estim_termino'] + timedelta(days=random.randint(0,3)),
            'estado': 'Almacén'
        })
    else:
        # Si está en proceso, generamos terminados solo si hay cantidad terminada > 0
        cantidad_terminada = row['cantidad_total'] - row['cantidad_en_proceso']
        if cantidad_terminada > 0:
            productos_terminados.append({
                'id_terminado': f'PT-{row["id_proceso"].split("-")[1]}',
                'producto': row['producto'],
                'lote': row['lote'],
                'linea_produccion': row['linea_produccion'],
                'cantidad_terminada': cantidad_terminada,
                'fecha_termino': row['fecha_estim_termino'] + timedelta(days=random.randint(0,3)),
                'estado': 'Almacén'
            })

df_terminados = pd.DataFrame(productos_terminados)

# Guardar CSV si querés
df_wip.to_csv('productos_en_proceso.csv', index=False)
df_terminados.to_csv('productos_terminados.csv', index=False)
df_wip
df_terminados
```

## Algoritmo VIII-control de calidad

```python
import numpy as np
import random

# Semilla para reproducibilidad
random.seed(42)
np.random.seed(42)

def simular_control_calidad(df_terminados):
    resultados_qc = []

    for idx, fila in df_terminados.iterrows():
        # Probabilidad de aprobar (por ejemplo, 85%)
        aprobado = np.random.choice([True, False], p=[0.85, 0.15])

        if aprobado:
            porcentaje_defectos = round(random.uniform(0, 2), 2)  # muy bajo defectos si aprueba
            observacion = 'Sin observaciones relevantes'
            estado_qc = 'Aprobado'
        else:
            porcentaje_defectos = round(random.uniform(5, 20), 2)  # defectos significativos
            observacion = random.choice([
                'Defectos en pintura',
                'Problemas de ensamblaje',
                'Daños en superficie',
                'Dimensiones fuera de tolerancia'
            ])
            estado_qc = 'Rechazado'

        resultados_qc.append({
            'id_terminado': fila['id_terminado'],
            'producto': fila['producto'],
            'lote': fila['lote'],
            'fecha_inspeccion': fila['fecha_termino'] + pd.Timedelta(days=random.randint(1,5)),
            'estado_qc': estado_qc,
            'porcentaje_defectos': porcentaje_defectos,
            'observacion': observacion
        })

    return pd.DataFrame(resultados_qc)

# Simular QC
df_control_calidad = simular_control_calidad(df_terminados)

df_control_calidad.head()
```

```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta

# Inicializar Faker y semillas
fake = Faker('es_AR')
random.seed(101)
np.random.seed(101)
Faker.seed(101)

# Parámetros
num_inmuebles = 30

tipos_inmuebles = [
    "Oficina", "Depósito", "Terreno", "Planta industrial",
    "Local comercial", "Edificio", "Galpón", "Centro logístico"
]

estados_inmueble = ["Operativo", "Arrendado", "Mantenimiento", "Disponible", "Inactivo"]
ubicaciones = ["CABA", "Gran Buenos Aires", "Córdoba", "Mendoza", "Rosario", "Neuquén", "Salta"]

# Generar datos
inmuebles = []

for i in range(num_inmuebles):
    tipo = random.choice(tipos_inmuebles)
    direccion = fake.address().replace("\n", ", ")
    ciudad = random.choice(ubicaciones)
    estado = random.choices(estados_inmueble, weights=[0.5, 0.2, 0.1, 0.1, 0.1])[0]
    fecha_adquisicion = fake.date_between(start_date='-25y', end_date='-2y')
    valor_adquisicion = round(random.uniform(1.0, 150.0), 2)
    superficie_m2 = round(random.uniform(100.0, 5000.0), 2)
    id_inmueble = f"INM-{1000 + i}"

    inmuebles.append({
        "id_inmueble": id_inmueble,
        "tipo_inmueble": tipo,
        "direccion": direccion,
        "ciudad": ciudad,
        "estado": estado,
        "fecha_adquisicion": fecha_adquisicion,
        "valor_adquisicion": valor_adquisicion,
        "superficie_m2": superficie_m2
    })

# Crear DataFrame
df_inmuebles = pd.DataFrame(inmuebles)

# Guardar como CSV si lo necesitás
df_inmuebles.to_csv("inventario_inmuebles.csv", index=False)

# Mostrar ejemplo
df_inmuebles
```
